### **Conexión a datos**

La conexión a los datos se realiza por medio de un *request*, donde para este caso se necesita usar una URL especifíca para determinar si se desea obtener la información entre taxis, rutas regulares o servicios especiales. A continuación se muestra un ejemplo:

```python
import requests

# Clave de acceso a la API
API_KEY = "TU_API_KEY"

# Endpoint del dataset de servicios de taxi
url = (
    "https://consejodetransportepublico.cloudapi.junar.com/"
    "api/v2/datastreams/SERVI-DE-TAXIS-POR-CLASE/data.json/"
    f"?auth_key={API_KEY}&limit=10000"
)

# Realizar solicitud y conversión a JSON
response = requests.get(url)

print("HTTP Status:", response.status_code)

data = response.json()
```

Sin embargo, es importante revisar los parametros posibles para el request, ya que para este caso que se usa **Taxis**, por defecto solo devuelve información de las placas TA (Alajuela), por lo que para obtener la información completa sería necesario realizar un ciclo para compilar toda esta información. ### **Conexión a datos**

La conexión a los datos se hace por medio de un `request` a través de Python. Es importante saber que el servicio WFS (WebFeatureService) tiene la capacidad de almacenar distintas capas de información, por lo que es importante saber cual es la capa deseada, en este caso sería `IGN_25:edificacionesprominentes_25k`. A continuación un ejemplo de la conexión.

```python
WFS_URL_EDIFICACIONES = "https://geos.snitcr.go.cr/be/IGN_25/wfs"
LAYER_EDIFICACIONES = "IGN_25:edificacionesprominentes_25k"

url = (
    f"{WFS_URL_EDIFICACIONES}"
    "?service=WFS"
    "&version=2.0.0"
    "&request=GetFeature"
    f"&typeName={LAYER_EDIFICACIONES}"
    "&outputFormat=application/json"
)

gdf_edificaciones = gpd.read_file(url)
```
