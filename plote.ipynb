```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
gdf_poblacion = gpd.read_file("datos/poblacion_distrito.geojson")
```

```python
print("Shape:", gdf_poblacion.shape)

print("\nColumnas:")
print(gdf_poblacion.columns.tolist())

print("\nInfo:")
gdf_poblacion.info()

print("\nValores nulos:")
display(gdf_poblacion.isna().sum())

print("\nValores únicos:")
display(gdf_poblacion.nunique())

print("\nHead:")
display(gdf_poblacion.head())
```

```python
gdf_poblacion = gdf_poblacion.drop(columns=[
    "id",
    "value",
    "source",
    "qualitative_value",
    "notes",
    "geometry"
])
```

```python
print("codigo_pcd")
print("Mínimo:", gdf_poblacion["codigo_pcd"].min())
print("Máximo:", gdf_poblacion["codigo_pcd"].max())

print("\nnom_es")
print("Mínimo:", gdf_poblacion["nom_es"].min())
print("Máximo:", gdf_poblacion["nom_es"].max())

print("\nlabel_value")
print("Mínimo:", gdf_poblacion["label_value"].min())
print("Máximo:", gdf_poblacion["label_value"].max())
```

```python
print("\nHead:")
display(gdf_poblacion.head())
```

```python
gdf_poblacion.to_csv(
    "poblacion_distritos.csv",
    index=False,
    encoding="utf-8-sig"
)
```
