```python
import requests
import pandas as pd

url = "https://datos.aresep.go.cr/ws.datosabiertos/Services/IT/Autobus.svc/ObtenerEstadisticasAutobuses/2026/4"

response = requests.get(url)

print("Status HTTP:", response.status_code)

data = response.json()

print("Tipo de respuesta:", type(data))
print("Llaves:", data.keys())

df_autobus = pd.DataFrame(data["value"])

print("Shape:", df_autobus.shape)

print("\nColumnas:")
print(df_autobus.columns.tolist())

print("\nInfo:")
df_autobus.info()

print("\nHead:")
display(df_autobus.head())
```

```python

dfs = []

def junar_to_dataframe(data):
    result = data.get("result", {})

    if result.get("fType") == "ERROR":
        print("Error interno:")
        print(result.get("fStr"))
        return None

    if "fArray" not in result:
        print("No se encontró fArray.")
        return None

    items = result["fArray"]

    headers = [
        item["fStr"]
        for item in items
        if item.get("fHeader") is True
    ]

    values = [
        item["fStr"]
        for item in items
        if not item.get("fHeader")
    ]

    rows = [
        values[i:i + len(headers)]
        for i in range(0, len(values), len(headers))
    ]

    return pd.DataFrame(rows, columns=headers)


def consultar_clase_taxi(clase):
    url = (
        "https://consejodetransportepublico.cloudapi.junar.com/"
        "api/v2/datastreams/SERVI-DE-TAXIS-POR-CLASE/data.json/"
        f"?auth_key={API_KEY}"
        "&limit=1000000"
        f"&clase-de-placa={clase}"
    )

    print(f"Consultando {clase}...")

    try:
        response = requests.get(url, timeout=300)
        print("HTTP:", response.status_code)

        if response.status_code != 200:
            print(response.text[:500])
            return None

        data = response.json()
        df_temp = junar_to_dataframe(data)

        if df_temp is None:
            return None

        df_temp["clase_placa_consultada"] = clase
        print("Registros obtenidos:", len(df_temp))

        return df_temp

    except Exception as e:
        print("Error en la consulta:")
        print(e)
        return None
```
